```python
import pandas as pd
import os


path = os.path.join(os.getcwd(),'data','train_set','power--2008-08-22_2010-07-10.csv')
target = pd.read_csv(path, nrows=5, index_col=0)

print target.head(5)
```

```python
path = os.path.join(os.getcwd(),'data','train_set','context--2008-08-22_2010-07-10--dmop.csv')
dmop = pd.read_csv(path, nrows=5, index_col=0)
dmop.columns = ['dmop']

print dmop.head(5)
```

```python
import datetime

data = pd.concat([target,dmop], axis=1, join='outer', ignore_index=False)
data.index = [datetime.datetime.fromtimestamp(ms/1000.0) for ms in data.index]
data.sort_index(inplace=True)

print data.head()
```
